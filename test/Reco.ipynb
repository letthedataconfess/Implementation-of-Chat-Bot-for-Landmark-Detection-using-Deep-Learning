```python
#Lets read the BX-Books.csv file into a data frame
bookFile="C:\\RahulSureshGhali\\work\\MyCode\\MachineLearning\\NNRecommendations\\BX-Books.csv"
books=pd.read_csv(bookFile,sep=";",header=0,error_bad_lines=False, usecols=[0,1,2],index_col=0,names=['isbn',"title","author"],encoding="ISO-8859-1")
```

```python
#Lets see the top five rows from the dataframe
books.head()
```

```python
#Lets write a function called bookMeta which accepts a isbn number and returns the title and author of the book
def bookMeta(isbn):
    title = books.at[isbn,"title"]
    author = books.at[isbn,"author"]
    return title, author
```

```python
#Lets retain ratings only for those books which are present in the books dataframe.
ratings= ratings[ratings["isbn"].isin(books.index)]
```

```python
#Lets write a function faveBooks which will return the favourite books for a given user
def faveBooks(user,N):
    userRatings = ratings[ratings["user"]==user]
    sortedRatings = pd.DataFrame.sort_values(userRatings,['rating'],ascending=[0])[:N] 
    sortedRatings["title"] = sortedRatings["isbn"].apply(bookMeta)
    return sortedRatings
```

```python
#Lets create a ratings matrix where users are represented as rows and books as columns
usersBooksRatingMatrix=pd.pivot_table(ratings, values='rating',index=['user'], columns=['isbn'])
```

```python
#Lets get all ratings for user1 and see the top 5 records
user1Ratings = usersBooksRatingMatrix.transpose()[user1]
user1Ratings.head()
```

```python
#Lets get all ratings for user1 and see the top 5 records
user2Ratings = usersBooksRatingMatrix.transpose()[user2]
user2Ratings.head()
```

```python
#Now lets figure out the hamming distance between these 2 users
from scipy.spatial.distance import hamming 
hamming(user1Ratings,user2Ratings)
```

```python
#Now lets write a function called distance which computes the hamming distance between 2 users 
import numpy as np
def distance(user1,user2):
        try:
            user1Ratings = usersBooksRatingMatrix.transpose()[user1]
            user2Ratings = usersBooksRatingMatrix.transpose()[user2]
            distance = hamming(user1Ratings,user2Ratings)
        except: 
            distance = np.NaN
        return distance
```

```python
#Lets test the function
distance(user1,user2)
```

```python
#Lets say we need to compute the k nearest neighbors for a user "204622". Lets call this user as a activeUser
#In order to do this we will have to compute the distance of this user with every other users in the dataset
activeUser = 204622
#Lets get all users from the matrix except the activeUser
allUsers = pd.DataFrame(usersBooksRatingMatrix.index)
allUsers = allUsers[allUsers.user!=activeUser]
allUsers.head()
```

```python
#Lets add a column distance to this dataframe and lets compute the distance of the activeUser from each user
allUsers["distance"] = allUsers["user"].apply(lambda x: distance(activeUser,x))
allUsers.head()
```

```python
#Lets now identify the K nearest neighbors from the active user by sorting allUsers based on the distance 
#from the active user and selecting the first 10 users 
K = 10
KnearestUsers = allUsers.sort_values(["distance"],ascending=True)["user"][:K]
KnearestUsers
```

```python
#Lets put this together in a function. Lets write a function nearestNeighbors which accepts a user and 
#a value for 'K' which represents the number of nearest neighbors to be identified for that user.
#The default value for 'K' would be 10 
def nearestNeighbors(activeUser,K=10):
    allUsers = pd.DataFrame(usersBooksRatingMatrix.index)
    allUsers = allUsers[allUsers.user!=activeUser]
    allUsers["distance"] = allUsers["user"].apply(lambda x: distance(activeUser,x))
    KnearestUsers = allUsers.sort_values(["distance"],ascending=True)["user"][:K]
    return KnearestUsers
```

```python
#Lets test this function
KnearestUsers = nearestNeighbors(activeUser)
KnearestUsers
```

```python
#Lets get the nearest neigbors ratings from the ratings matrix
NNRatings = usersBooksRatingMatrix[usersBooksRatingMatrix.index.isin(KnearestUsers)]
NNRatings
```

```python
#Let us compute the average ratings for each and every book
#The np.nanmean will compute the average of ratings and will ignore all NaN's
#Finally for some of the books the average ratings will be NaN as those books might not have been 
#rated by any of the user. We need to drop such books by applying the dropna() function
avgRating = NNRatings.apply(np.nanmean).dropna()
avgRating.head()
```

```python
#From the ratings matrix lets get the books which the active user has already read
#The dropna function will drop the books which the user has not rated
booksAlreadyRead = usersBooksRatingMatrix.transpose()[activeUser].dropna().index
booksAlreadyRead
```

```python
#Lets discard the books from the avgRatings dataset, which have been allready read/rated by the active user
avgRating = avgRating[~avgRating.index.isin(booksAlreadyRead)]
```

```python
#Now lets us find out top 3 books from the avgRatings dataset
N=3
topNBooks = avgRating.sort_values(ascending=False).index[:N]
```

```python
#Finally lets apply the bookMeta function to get the name and author of each of these books
pd.Series(topNBooks).apply(bookMeta)
```

```python
#Lets put this together in a function
#Lets write a function topNRecommendations which accepts a user and number of recommendations it should return
def topN(activeUser,N=3):
    KnearestUsers = nearestNeighbors(activeUser)
    NNRatings = usersBooksRatingMatrix[usersBooksRatingMatrix.index.isin(KnearestUsers)]
    avgRating = NNRatings.apply(np.nanmean).dropna()
    booksAlreadyRead = usersBooksRatingMatrix.transpose()[activeUser].dropna().index
    avgRating = avgRating[~avgRating.index.isin(booksAlreadyRead)]
    topNISBNs = avgRating.sort_values(ascending=False).index[:N]
    return pd.Series(topNISBNs).apply(bookMeta)
```

```python
#Now lets get the top 10 recommendations for this user
topN(204813,10)
```
